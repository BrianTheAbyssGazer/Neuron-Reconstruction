```python
import glfw3
from PIL import Image
import numpy as np
import yt
img = Image.open('OneDrive/Documents/nTracer sample.tif')
h, w = np.shape(img)
nframes = int(img.n_frames/4)
if nframes%4!=0:
    raise exception("file type should be tiff with RGBA channels")
img_r = np.zeros((h, w, nframes))
img_g = np.zeros((h, w, nframes))
img_b = np.zeros((h, w, nframes))
img_a = np.zeros((h, w, nframes))
for i in range(nframes-1):
	img.seek(i*4+1)
	img_r[:, :, i] = np.array(img)
	img.seek(i * 4 + 2)
	img_g[:, :, i] = np.array(img)
	img.seek(i * 4 + 3)
	img_b[:, :, i] = np.array(img)
	img.seek(i * 4 + 4)
	img_a[:, :, i] = np.array(img)

img_r = img_r/np.amax(img_r)
img_g = img_g/np.amax(img_g)
img_b = img_b/np.amax(img_b)
img_a = img_a/np.amax(img_a)
```

```python
data = dict(density = (img_r, "g/cm**3"))
ds = yt.load_uniform_grid(data, img_r.shape)
yt.interactive_render(ds)
```

```python
import matplotlib.pyplot as plt
imgplot = plt.imshow(img_r[:,:,100])
plt.show()
```
